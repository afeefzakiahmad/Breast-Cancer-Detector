```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

## Creating the model

```python
class MyModel:
  def __init__(self, X):
    self.weights = torch.rand((X.shape[1],1), requires_grad = True, dtype = X.dtype)
    self.bias = torch.zeros(1, requires_grad = True, dtype =X.dtype)

  def forward(self, X):
    z_pred = torch.matmul(X, self.weights) + self.bias
    y_pred = torch.sigmoid(z_pred)
    return y_pred

  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)


    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
model = MyModel(X_train)
```

```python
learning_rate = 0.1
epochs = 10000
```

```python
# Loop
for epoch in range(epochs):

  # Forward Pass
  y_pred = model.forward(X_train)

  # Loss Calculation
  loss = model.loss_function(y_pred, y_train)

  # Backprop
  loss.backward()

  # Update Parameters
  with torch.no_grad():
    model.weights -= learning_rate * model.weights.grad
    model.bias -= learning_rate * model.bias.grad

  # Zero Gradient
  model.weights.grad.zero_()
  model.bias.grad.zero_()

  if epoch % 1000 == 0:
    print(f'{epoch=} , loss = {loss.item()}')
```

```python
with torch.no_grad():
  y_pred = model.forward(X_test)
  y_threshold= (y_pred>0.5).float()
  accuracy = (y_threshold == y_test).float().mean()
  print(f'Test accuracy: {accuracy.item()}')
```

```python
class Node:
  def __init__(self,input_size, sigmoid = False):
    self.weights = torch.rand((input_size,1), requires_grad = True, dtype = torch.float64)
    self.bias = torch.zeros(1, requires_grad = True, dtype = torch.float64)
    self.sigmoid = sigmoid
  def forward(self, a_prev):
    z = torch.matmul(a_prev, self.weights) + self.bias
    if self.sigmoid:
      a = torch.sigmoid(z)
    else:
      a = torch.relu(z)
    return a
```

```python
class BigModel:
  def __init__(self):

    # Layer 1 (ReLU activation)
    layer1_node1 = Node(30)
    layer1_node2 = Node(30)

    # Layer 2 (sigmoid activation)
    layer2_node1 = Node(2, sigmoid = True)

    self.layer1_node1 = layer1_node1
    self.layer1_node2 = layer1_node2
    self.layer2_node1 = layer2_node1
    self.num_nodes = 3
    self.num_layers = 2

  def forward(self, X_train):
    a11 = self.layer1_node1.forward(X_train)
    a12 = self.layer1_node2.forward(X_train)
    a1 = torch.cat((a11,a12), dim = 1)

    a2 = self.layer2_node1.forward(a1)
    return a2


  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)


    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss
```

```python
bigmodel = BigModel()
```

```python
learning_rate = 0.1
epochs = 10000
```

```python
# Loop
for epoch in range(epochs):

  # Forward Pass
  y_pred = bigmodel.forward(X_train)

  # Loss Calculation
  loss = bigmodel.loss_function(y_pred, y_train)

  # Backprop
  loss.backward()

  # Update Parameters
  with torch.no_grad():
    bigmodel.layer1_node1.weights -= learning_rate * bigmodel.layer1_node1.weights.grad
    bigmodel.layer1_node1.bias -= learning_rate * bigmodel.layer1_node1.bias.grad
    bigmodel.layer1_node2.weights -= learning_rate * bigmodel.layer1_node2.weights.grad
    bigmodel.layer1_node2.bias -= learning_rate * bigmodel.layer1_node2.bias.grad
    bigmodel.layer2_node1.weights -= learning_rate * bigmodel.layer2_node1.weights.grad
    bigmodel.layer2_node1.bias -= learning_rate * bigmodel.layer2_node1.bias.grad

  # Zero Gradient
  bigmodel.layer1_node1.weights.grad.zero_()
  bigmodel.layer1_node1.bias.grad.zero_()
  bigmodel.layer1_node2.weights.grad.zero_()
  bigmodel.layer1_node2.bias.grad.zero_()
  bigmodel.layer2_node1.weights.grad.zero_()
  bigmodel.layer2_node1.bias.grad.zero_()
  if epoch % 1000 == 0:
    print(f'Epoch = {epoch+1} , loss = {loss.item()}')
```

```python
with torch.no_grad():
  y_pred = bigmodel.forward(X_test)
  y_threshold= (y_pred>0.5).float()
  accuracy = (y_threshold == y_test).float().mean()
  print(f'Test accuracy: {accuracy.item()}')
```

# Model with 4 layers

```python
class Layer:
  def __init__(self, input_features, num_nodes, sigmoid = False):
    self.weights = ((torch.rand((num_nodes, input_features), dtype = torch.float64)) - 0.5 ).detach().clone().requires_grad_(True)
    self.bias = torch.zeros(num_nodes, requires_grad = True, dtype = torch.float64)
    self.sigmoid = sigmoid

  def forward(self, a_prev):
    z = torch.matmul(a_prev, torch.transpose(self.weights,0,1)) + self.bias
    if self.sigmoid:
      a = torch.sigmoid(z)
    else:
      a = torch.relu(z)
    return a
```

```python
class HugeModel:
  def __init__(self, input_features):

    # Layer 1
    self.layer1 = Layer(input_features, 15)
    self.layer2 = Layer(15, 8)
    self.layer3 = Layer(8, 4)
    self.layer4 = Layer(4, 1 , sigmoid = True)

  def forward(self, X_train):
    a1 = self.layer1.forward(X_train)
    a2 = self.layer2.forward(a1)
    a3 = self.layer3.forward(a2)
    a4 = self.layer4.forward(a3)
    return a4

  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)


    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss

  def predict(self, X):
    with torch.no_grad():
      y_pred = self.forward(X)
      y_threshold= (y_pred>0.5).int()
      return y_threshold
```
