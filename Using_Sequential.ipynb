```python
df.drop(columns=['id','Unnamed: 32'], inplace=True)
df.head()
```

### Train Test split

```python
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,1:],df.iloc[:,0], test_size = 0.2)
```

### Scaling and encoding

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train = torch.from_numpy(X_train).float()
X_test = torch.from_numpy(X_test).float()
y_train = torch.from_numpy(y_train).float()
y_test = torch.from_numpy(y_test).float()
```

## Creating Model

```python
class MyModel(nn.Module):
  def __init__(self, input_features):
    super().__init__()
    self.network = nn.Sequential(
        nn.Linear(input_features, 15),
        nn.ReLU(),
        nn.Linear(15, 8),
        nn.ReLU(),
        nn.Linear(8, 4),
        nn.ReLU(),
        nn.Linear(4, 1),
        nn.Sigmoid()
    )

  def forward(self, X):
    out = self.network(X)
    return out


  def loss_function(self, y_pred, y):
    epsilon = 1e-7
    y_pred = torch.clamp(y_pred, epsilon, 1 - epsilon)


    loss = -(y * torch.log(y_pred) + (1 - y) * torch.log(1 - y_pred)).mean()
    return loss

  def predict(self, X):
    with torch.no_grad():
      y_pred = self.forward(X)
      y_pred = (y_pred > 0.5).float()
      return y_pred
```

```python
model = MyModel(X_train.shape[1])
```

```python
learning_rate = 10
epochs = 1000
```

```python
model.network[0].weight.shape
```

```python
y_before = model.predict(X_train)
y_before.sum()
```

```python
# Loop
for epoch in range(epochs):
  y_pred = model(X_train)
  loss = model.loss_function(y_pred, y_train)
  loss.backward()

  with torch.no_grad():
    for param in model.parameters():
      param -= learning_rate * param.grad
      param.grad.zero_()

  if epoch % 100 == 0:
    print(f"Epoch: {epoch} Loss: {loss}")
```

```python
z1 = torch.matmul(X_train,torch.transpose(model.network[0].weight, 1, 0)) + model.network[0].bias
a1 = torch.relu(z1)
z2 = torch.matmul(a1, torch.transpose(model.network[2].weight, 1, 0)) + model.network[2].bias
a2 = torch.relu(z2)
z3 = torch.matmul(a2, torch.transpose(model.network[4].weight, 1, 0)) + model.network[4].bias
a3 = torch.relu(z3)
z4 = torch.matmul(a3, torch.transpose(model.network[6].weight, 1, 0)) + model.network[6].bias
a4 = torch.sigmoid(z4)
```

```python
y_after = model.predict(X_train)
print(f'{y_after.sum()=} , {y_train.sum()=}')
```

```python
y_compare = (y_after == y_train).float()
print(f'{y_compare.sum()=} , {y_train.shape[0]=}')
```
